SUM() as the aggregate function

```python
%%sql

SELECT state, sum(gift_amount) as total FROM cube GROUP BY state;
```

```python
%%sql

SELECT city, state, sum(gift_amount) as total FROM cube GROUP BY city ORDER BY state, city ASC;
```

```python
%%sql

SELECT college, city, state, sum(gift_amount) as total FROM cube GROUP BY college, city ORDER BY state, city, college ASC;
```

```python
%%sql

SELECT college, city, state, sum(gift_amount) as total FROM cube GROUP BY college, city ORDER BY total DESC;
```

```python
%%sql

select sum(gift_amount) from cube where state="CO";
```

```python
%%sql

select * from cube;
```

```python
%%sql

select count(DISTINCT state), count(DISTINCT city), count(DISTINCT college), count(DISTINCT allocation_subcategory), count(DISTINCT gift_allocation), count(DISTINCT major),
count(DISTINCT state) + count(DISTINCT city) + count(DISTINCT college) + count(DISTINCT allocation_subcategory) + count(DISTINCT gift_allocation) + count(DISTINCT major) as total
from cube;
```
